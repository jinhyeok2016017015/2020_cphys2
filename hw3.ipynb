Boston House Price

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
boston = load_boston()
X=[[i] for i in boston.data[:,12]]
y=[[i] for i in boston.target]
reg=LinearRegression()
reg.fit(X,y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(boston.data[:,12],boston.target)
plt.plot(X,reg.predict(X),'k-')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.show()
```

결론 : LSTAT가 높을수록 주인이 거주하는 주택 가격의 중간값이 낮아진다.
